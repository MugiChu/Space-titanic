```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
inf = df.select_dtypes([np.number]).columns
```

```python
inf
```

```python
fig, axs = plt.subplots(6,figsize=(30,30))
for idx, name in enumerate(inf):
    df[inf[idx]].plot(ax=axs[idx])
    axs[idx].set_title(name, fontsize=16)
plt.tight_layout()
plt.show()
```
